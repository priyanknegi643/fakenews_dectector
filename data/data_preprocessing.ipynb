```python
# remove unnecessary symbols
def clean_text(text):
    text = text.lower()
    text = re.sub(r'^.*?\(.*?\)\s?-\s?', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r"http\S+", "", text)   # remove URLs
    text = re.sub(r"\s+", " ", text)      # remove extra spaces
    return text

df["content"] = df["content"].apply(clean_text)
```
